You are given an m x n matrix board containing letters 'X' and 'O', capture regions that are surrounded:

- Connect: A cell is connected to adjacent cells horizontally or vertically.
- Region: To form a region connect every 'O' cell.
- Surround: The region is surrounded with 'X' cells if you can connect the region with 'X' cells and none of the region cells are on the edge of the board.

To capture a surrounded region, replace all 'O's with 'X's in-place within the original board. You do not need to return anything.

```python
from typing import List

class Solution:
    def solve(self, board: List[List[str]]) -> None:
        n_rows, n_cols = len(board), len(board[0])

        # 1st solution using BFS - 0ms Beats 100%
        """
        def bfs(r,c):
            q = deque([(r,c)])
            board[r][c] = "sO"

            while q:
                x, y = q.popleft()
                for dx, dy in [(1,0), (-1,0), (0,1), (0,-1)]:
                    nx, ny = x+dx, y+dy
                    if 0 <= nx < n_rows and 0 <= ny < n_cols and board[nx][ny] == 'O':
                        board[nx][ny] = 'sO'
                        q.append((nx, ny))


        for r in range(n_rows):
            if board[r][0]        == "O": bfs(r,0)
            if board[r][n_cols-1] == "O": bfs(r,n_cols-1)
        
        for c in range(n_cols):
            if board[       0][c] == "O": bfs(0,c)
            if board[n_rows-1][c] == "O": bfs(n_rows-1,c)
        """

        # 2nd solution using DFS - 3ms Beats 86.45%
        # """
        o_visited = set()
        def dfs(r,c):
            board[r][c] = "sO"
            o_visited.add((r,c))

            for dr, dc in [(1,0), (-1,0), (0,1), (0,-1)]:
                connect_r, connect_c = r+dr, c+dc
                if 0 <= connect_r < n_rows and 0 <= connect_c < n_cols \
                    and board[connect_r][connect_c] == 'O' and (connect_r, connect_c) not in o_visited:
                    dfs(connect_r, connect_c)

        for r in range(n_rows):
            if board[r][0]        == "O": dfs(r,0)
            if board[r][n_cols-1] == "O": dfs(r,n_cols-1)
        
        for c in range(n_cols):
            if board[       0][c] == "O": dfs(0,c)
            if board[n_rows-1][c] == "O": dfs(n_rows-1,c)

        # """
        

        # Shared code between the two solutions
        for r in range(n_rows):
            for c in range(n_cols):
                if board[r][c] == "O":
                    board[r][c] = "X"
                elif board[r][c] == "sO": # S <-> still
                    board[r][c] = "O"
```
