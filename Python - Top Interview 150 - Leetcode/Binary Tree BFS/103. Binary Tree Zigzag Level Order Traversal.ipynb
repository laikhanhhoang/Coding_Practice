Given the root of a binary tree, return the zigzag level order traversal of its nodes' values. (i.e., from left to right, then right to left for the next level and alternate between).

```python
from collections import deque
from typing import List, Optional

# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def zigzagLevelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        if not root: return []

        curr_depth = 0
        curr_depth_list = deque([root])
        size_curr_depth = 1
        res = []

        while curr_depth_list:
            temp = []
            for i in range(0, size_curr_depth):
                rightmost_node = curr_depth_list.popleft()
                temp.append(rightmost_node.val)

                if rightmost_node.right:
                    curr_depth_list.append(rightmost_node.right)
                
                if rightmost_node.left:
                    curr_depth_list.append(rightmost_node.left)

            if not curr_depth % 2: temp = temp[::-1]

            res.append(temp)
            size_curr_depth = len(curr_depth_list)
            curr_depth += 1

        return res      
```
