Given the root of a binary tree, return the level order traversal of its nodes' values. (i.e., from left to right, level by level).

```python
from collections import deque
from typing import List, Optional

# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        if not root: return []

        curr_depth = deque([root])
        size_curr_depth = 1
        res = []

        while curr_depth:
            temp = []
            for i in range(0, size_curr_depth):
                leftmost_node = curr_depth.popleft()
                temp.append(leftmost_node.val)
                
                if leftmost_node.left:
                    curr_depth.append(leftmost_node.left)

                if leftmost_node.right:
                    curr_depth.append(leftmost_node.right)

            res.append(temp)
            size_curr_depth = len(curr_depth)
        
        return res
```
