Given the root of a Binary Search Tree (BST), return the minimum absolute difference between the values of any two different nodes in the tree.



```python
from typing import Optional

# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution1:
    def getMinimumDifference(self, root: Optional[TreeNode]) -> int:
        bst_value_to_list = []

        def dfs(node):
            if not node:
                return
            dfs(node.left)
            bst_value_to_list.append(node.val)
            dfs(node.right)

        dfs(root)

        # bst_value_to_list đã tăng dần
        res = float("inf")
        for i in range(1, len(bst_value_to_list)):
            res = min(res, abs(bst_value_to_list[i] - bst_value_to_list[i-1]))
        
        return res
```

```python
class Solution2:
    def getMinimumDifference(self, root: Optional[TreeNode]) -> int:
        inorder = []
        res = float('inf')

        def dfs(root):
            if not root:
                return []
            return dfs(root.left) + [root.val] + dfs(root.right)
        inorder = dfs(root)

        for i in range(1, len(inorder)):
            res = min(res, abs(inorder[i]-inorder[i-1]))

        return res
```

```python
class Solution3:
    def getMinimumDifference(self, root: Optional[TreeNode]) -> int:
        prev = None      
        min_diff = float('inf')

        def inorder(node):
            if not node: return
            
            nonlocal prev, min_diff

            # Traverse left subtree
            inorder(node.left)
            # Process current node
            if prev is not None: min_diff = min(min_diff, node.val - prev)
            prev = node.val
            # Traverse right subtree
            inorder(node.right)

        inorder(root)
        return min_diff
```
